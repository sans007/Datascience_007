```python
numerical_features = [feature for feature in df.columns if df[feature].dtype != 'O']

print('Number of numerical variables: ', len(numerical_features))

df[numerical_features].head()
```

```python
for feature in numerical_features:
    df1 = df.copy()
    if 0 in df1[feature].unique():
      pass
    else:
      df1[feature] = np.log(df1[feature])
      df1.boxplot(column=feature)
      plt.ylabel(feature)
      plt.title(feature)
      plt.show()

```

```python

df['LUNG_CANCER'] = df['LUNG_CANCER'].replace(to_replace = {'YES' : 2 , 'NO': 1})
df['GENDER'] = df['GENDER'].replace(to_replace = {'M' : 2 , 'F': 1})
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['AGE'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (10, 10))
sns.countplot(y = 'AGE',hue='LUNG_CANCER', data =df, palette = 'viridis')
plt.title('death rate vary with age ')
plt.ylabel('age')
plt.legend(labels=['No death','death'])
plt.show()
```
