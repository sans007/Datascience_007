<br>1. Demonstrate three different methods for creating identical 2D arrays in NumPy. Provide the code for each
method and the final output after each method.


```python
array1 = np.array([[1, 2, 3], [4, 5, 6]])
print(array1)
print('Dimension-',np.ndim(array1))
```

```python
array2 = np.full((2, 3), 1)
array2[0] = [1, 2, 3]
array2[1] = [4, 5, 6]
print(array2)
print('Dimension-',np.ndim(array2))
```

```python
array3 = np.zeros((2, 3), dtype=int)
array3[0,:] = [1, 2, 3]
array3[1,:] = [4, 5, 6]
print(array3)
print('Dimension-',np.ndim(array3))
```

Shallow Copy:

A shallow copy creates a new object, but it does not create copies of objects that are referenced by the original object. Instead, it just copies the references to these objects.
This means that changes to mutable objects within the original object will affect the shallow copy and vice versa.

Deep Copy:

A deep copy creates a new object, and it recursively copies all objects found within the original object. This means that changes to the original object or its contents will not affect the deep copy and vice versa.

```python
random_array = np.random.randint(1, 10, (5, 6))

# Extract all even integers
even_integers = random_array[random_array % 2 == 0]

# Extract all odd integers
odd_integers = random_array[random_array % 2 != 0]

print("Original Array:\n", random_array)
print("\nEven Integers:\n", even_integers)
print("\nOdd Integers:\n", odd_integers)
```

tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.




```python

from scipy.stats import norm

# Create a NumPy array containing 1000 samples from a normal distribution
data = np.random.randn(1000)  # Standard normal distribution (mean=0, std=1)

# a) Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Histogram')

# b) Overlay a line plot representing the normal distribution's probability density function (PDF)
# Calculate the mean and standard deviation of the data
mean = np.mean(data)
std = np.std(data)

# Generate x values for the PDF
x_values = np.linspace(min(data), max(data), 100)

# Calculate the PDF for these x values
pdf_values = norm.pdf(x_values, mean, std)

# Plot the PDF
plt.plot(x_values, pdf_values, color='red', linestyle='-', label='Normal Distribution PDF')

# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot as 'Histogram with PDF Overlay'
plt.title('Histogram with PDF Overlay')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

<br>17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python

import seaborn as sns



np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)


def get_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Q1'
    elif x < 0 and y >= 0:
        return 'Q2'
    elif x < 0 and y < 0:
        return 'Q3'
    else:
        return 'Q4'


quadrants = np.array([get_quadrant(x_val, y_val) for x_val, y_val in zip(x, y)])


df = pd.DataFrame({
    'X': x,
    'Y': y,
    'Quadrant': quadrants
})


plt.figure(figsize=(8, 6))
scatter_plot = sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='Set1', style='Quadrant')


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')


plt.legend(title='Quadrants')


plt.show()
```

<br>18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'?

```python
sns.set(style='whitegrid')


x = np.linspace(0, 2 * np.pi, 500)
y = np.sin(x)


plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sine Wave', color='blue')


plt.grid(True)


plt.xlabel('X-axis (radians)')
plt.ylabel('Y-axis (amplitude)')


plt.title('Sine Wave Function')


plt.show()
```
