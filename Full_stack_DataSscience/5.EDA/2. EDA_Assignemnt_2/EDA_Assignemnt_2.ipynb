<br>1. What is the average rating for each product category?

```python
df['rating']=pd.to_numeric(df['rating'], errors = 'coerce')
```

```python
df['rating']
```

```python
df.info()
```

```python
average_rating = df.groupby('category')['rating'].mean().reset_index()
```

```python
average_rating
```

<br>2. What are the top rating_count products by category?

```python
df['rating_count'] = df['rating_count'].str.replace(",",'')
df['rating_count'] = df['rating_count'].astype('float64')
```

```python
df.columns
```

```python
average_rating_count1 = df.groupby(['product_id','product_name','category'])['rating_count'].mean().reset_index()
average_rating_count1
```

```python
average_rating_count2 = average_rating_count1.sort_values(by="rating_count", ascending=False)
average_rating_count2
```

```python


print("top rating_count products by category:")
for i in range(10):
    product_id = average_rating_count2.iloc[i]['product_id']
    product_name = average_rating_count2.iloc[i]['product_name']
    category = average_rating_count2.iloc[i]["category"]
    average_rating_count = average_rating_count2.iloc[i]["rating_count"]
    print((i+1) ,'.',product_id, ':',product_name,':', category,':', average_rating_count)
```

<br>3. What is the distribution of discounted prices vs. actual prices?

```python
df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python


plt.hist(data=df, x='discounted_price')
plt.hist(data=df, x = 'actual_price')

plt.show()
```

<br>4. How does the average discount percentage vary across categories?

```python

df['discount_percentage'] = df['discount_percentage'].str.replace("%",'')
df['discount_percentage'] = df['discount_percentage'].astype('float64')
```

```python
average_discount_percentage = df.groupby('category')['discount_percentage'].mean().reset_index()
average_discount_percentage
```

```python
sns.barplot(x=average_discount_percentage['category'], y=average_discount_percentage['discount_percentage'],color='g')
plt.xlabel("Category")
plt.ylabel("Average Discount Percentage")
plt.show()
```

<br>6.Among the top 5 most popular artists, which artist has the highest popularity on average?
 Calculate and display the average popularity for each artist.

```python
df2.columns
```

```python
popularity_on_average = df2.groupby('Artist')['Popularity'].mean().reset_index()
popularity_on_average
```

```python
Top_5_most_popular_artists = popularity_on_average.sort_values(by = 'Popularity', ascending = False )[:5]
Top_5_most_popular_artists
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Popularity', y='Artist', data= Top_5_most_popular_artists)
plt.title('average popularity for each artist')
plt.show()
```

<br>7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.
